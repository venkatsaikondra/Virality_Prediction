```python
import snscrape.modules.twitter as sntwitter
import pandas as pd

data = []

query = "crypto OR movies OR games lang:en since:2024-01-01"

for tweet in sntwitter.TwitterSearchScraper(query).get_items():

    data.append({
        "id": tweet.id,
        "text": tweet.content,
        "hashtags": len(tweet.hashtags) if tweet.hashtags else 0,
        "mentions": len(tweet.mentionedUsers) if tweet.mentionedUsers else 0,
        "length": len(tweet.content),
        "followers": tweet.user.followersCount,
        "following": tweet.user.friendsCount,
        "verified": int(tweet.user.verified),
    })

    if len(data) > 5000:
        break

df = pd.DataFrame(data)
df.to_csv("tweets_initial.csv", index=False)
```
